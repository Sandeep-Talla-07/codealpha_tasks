```python
# The fillna() function was used to fill the null values in dataset.
# There are many ways to fill the null values in dataset.
# In this case i am used bfill method and mode method to fill the missing values.
# bfill :- This method is used to fill the missing values with the next values.
# mode :- This method is used to fill the missing values with the mode of that column.
# Coming to the our dataset ''Gender', 'Married', 'Dependents', 'Self_Employed', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History' these columns contains missing values.
# This is not a good idea to drop the missing values.
# So i was filled the missing values with the different methods.

data['Gender'].fillna(data['Gender'].mode()[0], inplace=True)
data['Married'].fillna(data['Married'].mode()[0], inplace = True)
data['Dependents'].fillna(data['Dependents'].mode()[0], inplace = True)
data['Self_Employed'].fillna(data['Self_Employed'].mode()[0], inplace = True)
data['LoanAmount'].fillna(method = 'bfill', inplace = True)
data['Loan_Amount_Term'].fillna(method = 'bfill', inplace = True)
data['Credit_History'].fillna(data['Credit_History'].mode()[0], inplace=True)
```

# Machine Learning

```python
# The all below modules are imported from sklearn library.
# sklearn is a machine learning library.
# This library contains all machine learning algorithms and models.

from sklearn.model_selection import train_test_split # This module was used for split the dataset into train and test.
from sklearn.preprocessing import LabelEncoder # This module helps to encode the categorical columns into numerical columns.
from sklearn.ensemble import RandomForestClassifier # This module was used to build the Random Forest Classifier model.
from sklearn.metrics import accuracy_score, confusion_matrix # This module was used to find the accuracy and confusion matrix for model.
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, roc_curve # This module was used to find the precision, recall, f1 score, roc_auc_score, roc_curve for model.
```

```python
# Label Encoder was used to encode means convert the categorical columns into numerical columns.

label_encoder = LabelEncoder() # This an object for LabelEncoder().
char_columns = [col for col in data.columns if data[col].dtype == 'object'] # Storing the categorical columns in dataset.


for column in char_columns: # Iterate the categorical columns.
  data[column] = label_encoder.fit_transform(data[column]) # Encode the categorical columns into numerical columns.
                                                           #  It helps to build the model also on categorical values. 
```

```python
data_columns = [col for col in data.columns if data[col].dtype == 'object'] # Storing the categorical columns in dataset.
print(data_columns) # Print the categorical attributes in the dataset.
```

```python
data_num_columns = [col for col in data.columns if data[col].dtype != 'object'] # Storing the numerical columns in dataset.
print("Columns Names of the dataset : ",data_num_columns) # Print the numerical attributes in the dataset.

count = 0
# This loop was used to count the total number of columns in dataset
for i in enumerate(data_num_columns):
  count = count + 1
print("\nThe total number of columns present in the dataset : ",count) 
```

Model Building

```python
x = data.drop('Loan_Status', axis = 1) # Storing the all columns except Loan_Status columns in x.

y = data['Loan_Status'] # Storing the Loan_Status columns in y.

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 42, test_size = 0.2) # Splitting the dataset into train and test dataset.

model = RandomForestClassifier(random_state = 42) # Building the Random Forest Classifier model.

model.fit(x_train, y_train) # Fitting the model on train dataset.


y_train_pred = model.predict(x_train) # Predicting the model on train dataset.

y_test_pred = model.predict(x_test) # Predicting the model on test dataset.
```
